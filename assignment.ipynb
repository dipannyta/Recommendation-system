```python
import numpy as np
import pandas as pd


movie_details = pd.read_csv("movies_metadata.csv")
movie_details.head(10)
```

```python
import ast
movie_details['genres'] = movie_details['genres'].fillna('[]').apply(ast.literal_eval).apply(lambda g: [i['name'] for i in g] 
                                                                       if isinstance(g, list) else "")
```

```python
movie_details.head(10)
```

```python
movie_details['wr']=(movie_details['vote_count']/(movie_details['vote_count']+250))* movie_details['vote_average'] + (250/(movie_details['vote_count']+250))*7
```

```python
movie_details.head(10)
```

```python
gen = movie_details.apply(lambda g: pd.Series(g['genres']),axis=1).stack()
gen = gen.reset_index(level=1, drop=True)
gen.name = 'genre'
mov_d = movie_details.drop('genres', axis=1)
mov_d = mov_d.join(gen)
```

```python
mov_d.head(10)
```

```python
movie_details = movie_details.sort_values('wr', ascending=False).head(10)
```

```python
#top ten movies
movie_details.head(10)
```

```python
#take a genre and find movies of that genre, then build a table to display titlt, vote_count, vote_average, and  weighted rating
def genre_top_mov(genre):
    df = mov_d[mov_d['genre'] == genre]
    attribute = ['title', 'vote_count', 'vote_average', 'genre','wr']
    vote_counts = df[df['vote_count'].notnull()]['vote_count'].astype('int')
    vote_averages = df[df['vote_average'].notnull()]['vote_average'].astype('int')
    movie_details = df[(df['vote_count'] >= 250) 
                   & (df['vote_count'].notnull()) 
                   & (df['vote_average'].notnull())][attribute]
    movie_details['vote_count'] = movie_details['vote_count'].astype('int')
    movie_details['vote_average'] = movie_details['vote_average'].astype('int')
    
    movie_details['wr'] = movie_details.apply(lambda movie_details: 
                                                   (movie_details['vote_count']/(movie_details['vote_count']+250) * movie_details['vote_average']) + (250/(250+movie_details['vote_count'])* 7), axis=1)
    movie_details = movie_details.sort_values('wr', ascending=False).head(10)
    
    return movie_details
```

```python
#top ten movies for the genre : Animated
genre_top_mov('Animation').head(10)
```

```python
md = pd.read_csv("movies_metadata.csv")
md.head(10)
```

```python
#convert ids to int so that they can be used for opertion
def converttoint(num):
    try:
        x= int(num)
        return x
    except:
        return np.nan
```

```python
md['id'] = md['id'].apply(converttoint)
```

```python
md.head(10)
```

```python
link = pd.read_csv('links_small.csv')
link = link[link['tmdbId'].notnull()]['tmdbId'].astype('int')
```

```python
dff = md[md['id'].isin(link)]
```

```python
from scipy import stats
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel, cosine_similarity
```

```python
dff['tagline'] = dff['tagline'].fillna('')
dff['content'] = dff['overview'] + dff['tagline']
dff['content'] = dff['content'].fillna('')
```

```python
tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(dff['content'])
```

```python
tfidf_matrix.shape
```

```python
cosine_similarity = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
titles = dff['title']
indx = pd.Series(dff.index, index=dff['title'])
```

```python
def recommend(title):
    idx = indx[title]
    cs=enumerate(cosine_similarity[idx])
    similarity = list(cs)
    similarity = sorted(similarity, key=lambda x: x[1], reverse=True)
    movies = [i[0] for i in similarity]
    f=titles.iloc[movies]
    return f
```

```python
recommend('Toy Story').head(10)
```

```python
ratings = pd.read_csv('ratings_small.csv')
ratings.head(10)
```

```python
from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate
```

```python
# A reader object is created to parse a file containing ratings
reader = Reader(rating_scale=(1, 5))

data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)
```

```python
svd = SVD()
cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
from surprise.model_selection import train_test_split

trainset, testset = train_test_split(data, test_size=.25)

# Train the algorithm on the trainset, and predict ratings for the testset
svd.fit(trainset)
predictions = svd.test(testset)
```
